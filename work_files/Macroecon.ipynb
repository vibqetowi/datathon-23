```python
import numpy as np 
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
ROOT='./kaggle/input/dt23-test/'

#cpi_df = pd.read_csv(ROOT+'series/additional_data/us_CPI.csv', index_col=0)
inflation_df = pd.read_csv(ROOT+'series/additional_data/us_inflation_rate.csv', index_col=0)
m1_df = pd.read_csv(ROOT+'series/additional_data/M1SL.csv', index_col=0)

#cpi_df.index = pd.to_datetime(cpi_df.index)
inflation_df.index = pd.to_datetime(inflation_df.index)
m1_df.index = pd.to_datetime(m1_df.index) # Data is in millions


#print(cpi_df.head())
print( inflation_df.head())
print(m1_df.head())
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot cpi_df and inflation_df on ax1
#ax1.plot(cpi_df.index, cpi_df['CPI'], color='b', label='CPI')  # replace 'column_name' with actual column name
ax1.plot(inflation_df.index, inflation_df['inflation_rate (%)'], color='r', label='Inflation Rate')  # replace 'column_name' with actual column name
ax1.set_ylabel('Inflation Rate', color='black')
ax1.tick_params(axis='y', labelcolor='black')

# Create ax2 to share the same x-axis with ax1
ax2 = ax1.twinx()
ax2.plot(m1_df.index, m1_df['M1SL'], color='g', label='M1 Supply')  # replace 'column_name' with actual column name
ax2.set_ylabel('M1 Supply (in millions)', color='black')
ax2.tick_params(axis='y', labelcolor='black')

# Set title and show the plot with legend
plt.title('Inflation Rate, and M1 Supply Over Time')
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.95))
plt.show()
```
